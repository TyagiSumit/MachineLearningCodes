```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# color marker line width for axis, dataset and title, axis, gird

x = np.arange(0,5,0.1)
y = x**3
y2 = x**2
#plt.plot(x, y,color = "black", marker= "p")
plt.plot(x, y,color = "black",linewidth= 3,label = "x^3")
plt.plot(x, y2,color = "red",linewidth= 3, label = "x^2")

#plt.axis([0,10,0,200])

plt.grid()

plt.text(2, 80, "Text", fontsize= 12)

plt.ylabel("y")
plt.xlabel("x")
plt.legend()
plt.title("Matplotlib demo")

plt.show()
```
