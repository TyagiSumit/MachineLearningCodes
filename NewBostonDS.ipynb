```python
from sklearn import datasets
```

```python
boston = datasets.load_boston()
X = boston.data
Y = boston.target
X2 = boston.data
Y2 = boston.target
```

```python
X.shape
```

```python
import pandas as pd
df = pd.DataFrame(X)
print(boston.feature_names)
```

```python
df.columns = boston.feature_names
df.describe()
```

```python
for i in range(len(df.columns)):
    for j in range(i,len(df.columns)):
        name = df.columns[i] + "_" + df.columns[i]
        df[name] = df[df.columns[i]]*df[df.columns[j]]
```

```python
X = df.values
```

```python
df.describe()
```

```python
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y)
X2_train, X2_test, Y2_train, Y2_test = model_selection.train_test_split(X2,Y2)
```

```python
print(X_train.shape,X_test.shape)
print(Y_train.shape,Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
alg1 = LinearRegression()
alg2 = LinearRegression()
```

```python
alg1.fit(X_train, Y_train)
alg2.fit(X2_train, Y2_train)
```

```python
Y_pred = alg1.predict(X_test)
Y2_pred = alg2.predict(X2_test)
```

```python
#compare Y_pred and Y_test
import matplotlib.pyplot as plt
#plt.scatter(Y_pred,Y_tes|t)
plt.scatter(Y2_test,Y2_pred)
plt.scatter(Y_test,Y_pred)
plt.axis([0,40,0,40])
plt.show()
```

```python
print(alg1.score(X_test,Y_test))
print(alg2.score(X2_test,Y2_test))
```
