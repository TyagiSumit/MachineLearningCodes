```python
import numpy as np
from sklearn import linear_model,datasets
from sklearn import cross_validation,preprocessing
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data=np.genfromtxt("train_cpp.csv",delimiter=",")
print(data.shape)
X=data[:,:data.shape[1]-1]
print(X.shape)
Y=data[:,data.shape[1]-1:data.shape[1]]
a=np.ones(X.shape[0])
a=a.reshape(-1,1)
Y=Y.reshape(Y.shape[0],)
q=np.append(X,a,axis=1)
X=q
print(Y,Y.shape)
scale=preprocessing.StandardScaler()
scale.fit(X)
```

```python
# Spliting data
x_tr,x_ts,y_tr,y_ts=cross_validation.train_test_split(X,Y)
print(x_tr[2])
```

```python
clf=linear_model.LinearRegression()
clf.fit(x_tr,y_tr)
arr=clf.predict(x_ts)
print(score(arr,y_ts))
```

```python
def cost(y_p,y_ts):
    return ((y_p-y_ts)**2).mean()
def predict(a_y,x_ts):
    return np.dot(x_ts,a_y)


def gd(m,x_tr,y_tr,alpha):
 
    for i in range(m.shape[0]):
        for j in range(x_tr.shape[0]):
            m[i]+=alpha*(2*(y_tr[j]-(m*x_tr[j]).sum())*x_tr[j][i])/x_tr.shape[0]
            
            
    return
        
        
        

def run(x_tr,y_tr,alpha,itr):
    m=mf
    for i in range(itr):
        gd(m,x_tr,y_tr,alpha)
        print("Cost: ",cost(predict(m,x_tr),y_tr)," score: ",score(predict(m,x_tr),y_tr))
    return m


def score(y_p,y_ts):
    return 1-((cost(y_p,y_ts))/((y_ts-y_ts.mean())**2).mean())
    
    
```

```python
m=run(x_tr,y_tr,0.00001,100)
y_p=predict(m,x_ts)
print(score(y_p,y_ts))
```
